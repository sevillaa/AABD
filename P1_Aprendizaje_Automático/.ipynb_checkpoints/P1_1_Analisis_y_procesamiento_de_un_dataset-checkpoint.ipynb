```python
import pandas as pd

# Cargar el dataset titanic.csv
datos = pd.read_csv('titanic.csv')
filasPrevias = datos.shape[0]

# Eliminar filas duplicadas y luego las que contienen valores nulos
datos = datos.drop_duplicates().dropna()

filasPosteriores = datos.shape[0]
eliminadas = filasPrevias - filasPosteriores

datos.reset_index(drop=True, inplace=True)
# Guardar el dataset limpio en un archivo CSV
datos.to_csv('titanic_limpio.csv', index=False)

print('Filas eliminadas:', eliminadas)

```

1. CARGA DEL DATASET Y ANÁLISIS
    - Filas eliminadas -> 708 filas
    - Beneficios de eliminación -> La eliminación de filas duplicadas y con valores nulos mejora la calidad del dataset, reduce el ruido y evita posibles sesgos en el análisis y en los modelos de aprendizaje automático. 

2) ELEMENTOS QUE NO SON RELEVANTES:
    - Nombre (No tiene utilidad)
    - Ticket (El número asignado en el ticket no afecta a ningún resultado)
    - ID pasajero
    - Cabina en la que iba (con el dato de su clase es suficiente ya que las cabinas de cada clase estaban agrupadas)

```python
datos = datos.drop(columns=['Name', 'Ticket', 'Cabin', 'PassengerId'])
#Ya que considero que no son relevantes para el análisis de los datos

# Copia del dataset original para evitar cambios permanentes
datos_numericos = datos.copy()

# Convertir "Sex" a numérico (Label Encoding: Male -> 0, Female -> 1)
datos_numericos['Sex'] = datos_numericos['Sex'].map({'male': 0, 'female': 1})

# Convertir "Embarked" a variables numéricas (One-Hot Encoding)
datos_numericos = pd.get_dummies(datos_numericos, columns=['Embarked'], drop_first=True)

# Ahora, 'Embarked_C' y 'Embarked_Q' serán variables binarias (S es inferido -> Si C y Q son 0 embarcó en S)
```

```python
#Diagramas de dispersión
import matplotlib.pyplot as plt
import seaborn as sns

# Dibujar diagramas de dispersión
sns.pairplot(datos_numericos, hue='Survived')
plt.show()

# Calcular coeficientes de correlación
matriz_correlacion = datos_numericos.corr()
print(matriz_correlacion)

# Mostrar un mapa de calor de las correlaciones
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm')
plt.show()
```

- La relación más fuerte es entre sexo y supervivencia (0.53). Esto coincide con la política de evacuación que se empleó: "mujeres y niños primero". De la misma manera, la edad y la supervivencia tienen una correlación negativa, ya que a menor edad era más probable sobrevivir. Sin embargo, esta es más baja porque la edad y la clase tienen una correlación de -0.31, lo que indica que los pasajeros más jóvenes viajaban en las clases más bajas y, por ende, tenían menos posibilidades de alcanzar un bote salvavidas.

- Relación entre clase y tarifa (obvia).

- Puntos de embarque y clase:
    - Southampton tiene correlación positiva con la clase más baja.
    - Cherburgo tiene correlación más baja con la tercera clase y tarifas más altas, lo que sugiere que el mayor número de los de primera clase embarcaron allí.

- Familiares a bordo y tarifa:
    - Parch y Fare -> 0.389
    - SibSp y Fare -> 0.286
    Los pasajeros con más familiares a bordo pagaban tarifas más altas. Una posible interpretación es que había familias que viajaban juntas por ocio y, en aquella época, solo se lo podían costear las clases más altas, mientras que la gente que iba sola era gente que iba en busca de trabajo y, por tanto, con menos recursos.

#COMENTAR CON MAS DETALLE Y TODAS LAS COLUMNAS

```python
# ATRIBUTOS CATEGÓRICOS

# 1) Dibuja histogramas para los atributos categóricos
categorical_columns = ['Survived', 'Pclass', 'Sex', 'Embarked']

for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(datos[column], kde=False, bins=10)
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.show()

# 2) Diagramas de barras o de tartas para determinar las frecuencias de los valores de los atributos categóricos
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    datos[column].value_counts().plot(kind='bar')
    plt.title(f'Diagrama de barras de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.show()

    plt.figure(figsize=(10, 6))
    datos[column].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title(f'Diagrama de tarta de {column}')
    plt.ylabel('')
    plt.show()

# 3) Número de valores distintos para cada atributo y el valor más frecuente para cada atributo categórico
for column in categorical_columns:
    unique_values = datos[column].nunique()
    most_frequent_value = datos[column].mode()[0]
    print(f'Atributo: {column}')
    print(f'Número de valores distintos: {unique_values}')
    print(f'Valor más frecuente: {most_frequent_value}')
    print()

# 4) Atributo más balanceado y el menos balanceado
balance_info = {}
for column in categorical_columns:
    value_counts = datos[column].value_counts(normalize=True)
    balance_info[column] = value_counts.max() - value_counts.min()

most_balanced = min(balance_info, key=balance_info.get)
least_balanced = max(balance_info, key=balance_info.get)

print(f'Atributo más balanceado: {most_balanced}')
print(f'Atributo menos balanceado: {least_balanced}')
```

```python
#6. Determinar si hay outliers
#¿En que valores es mas esperable que haya outliers? -> En las columnas numéricas
#¿Cómo se pueden detectar empleando medidas de dispersión? -> Se pueden detectar mediante la desviación estándar
#¿Por qué? -> Porque la desviación estándar nos indica cuánto se alejan los valores de la media,
#               por lo que si un valor se aleja mucho de la media, es probable que sea un outlier

# Calcular la desviación estándar de los atributos numéricos
std_dev = datos_numericos.std()
mean = datos_numericos.mean()

# Crear un DataFrame con la media y la desviación estándar
# Esto no sería necesario ya que en el apartado 4 ya se calcula pero por trabajar con
#   los datos aislados y mas ordenados hemos preferido crear esta tabla
dispersion_df = pd.DataFrame({'Media': mean, 'Desviación Estándar': std_dev})

# Mostrar la tabla
print(dispersion_df)



```

- Edad -> 15.64, dispersión moderada, puede haber en los extremos de la distrbucion como 0.92 u 80 años 

- SibSp -> 0.644 dispersion baja

```python
#8. Normalizar y estandarizar el dataset

#Normalizar el dataset (datos_numericos)
from sklearn.preprocessing import StandardScaler, MinMaxScaler
columnas_numericas = datos_numericos.select_dtypes(include=["int64","float"]).columns

#Aplicar estandarización (StandardScaler)
scaler_std = StandardScaler()
datos_estandarizados = datos_numericos.copy()

datos_estandarizados[columnas_numericas] = scaler_std.fit_transform(datos_numericos[columnas_numericas])


#Aplicar la normalización (MinMaxScaler)
scaler_min_max = MinMaxScaler()
datos_normalizados = datos_numericos.copy()

datos_normalizados[columnas_numericas] = scaler_min_max.fit_transform(datos_numericos[columnas_numericas])

# Mostrar resultados
print("Dataset estandarizado:")
print(datos_estandarizados.head())

print("\nDataset normalizado:")
print(datos_normalizados.head())
```

```python
# CONCLUSIÓN
# La estandarización reduce el impacto de outliers al asjutarse en base a la desviación estándar sin ambargo no garantiza
# que los datos esten en un rango específico. Por otro lado la normalización es sensible a los outliers. Es por eso que este caso 
# la estandarización es la mejor opción ya que la normalización aparte de ser sensible a los outliers (presentes en este ejemplo) ,
# simplifica valores como la edad a [0,1] perdiendo variabilidad .
```
