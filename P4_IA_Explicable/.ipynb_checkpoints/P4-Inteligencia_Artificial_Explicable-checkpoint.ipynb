## Librerías a usar

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import precision_score, recall_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from shap import Explainer, initjs
from shap.plots import waterfall

from alibi.explainers import ALE, plot_ale

import lime.lime_tabular as tabular

import dice_ml as dice
from dice_ml.utils import helpers
```

## Semilla para los estados aleatorios

```python
SEED = 23895
POBLATION_SIZE = 3000
np.random.seed(SEED)
```

## Código remanente Práctica 3
Cabe destacar que acabamos utilizando un MLP compuesto de:
- Dos capas escondidas de 70 y 30 neuronas.
- Un coeficiente de regularización de 0.25
- Un coeficiente inicial de aprendizaje inicial de 0.25
- Una partición de los datos 55/45
- 5000 iteraciones máximas

```python
def plot_data(x1, x2, y, label1, label2, title) -> None:
    """ Realizamos dos dispersiones de puntos en función de la clase a la que pertenecen los datos, partiendo de las variables x1 y x2.

    Args:
        x1 (array_like): variable independiente 1
        x2 (array_like): variable independiente 2
        y (array_like): variable dependiente (clase)
        label1 (str): nombre de la variable independiente 1
        label2 (str): nombre de la variable independiente 2
        title (str): título del gráfico
    """
    plt.scatter(x1[y == 1], x2[y == 1], c='black', marker='x', label='Alto Riesgo (1)',)
    plt.scatter(x1[y == 0], x2[y == 0], c='red', marker='o', label='Bajo Riesgo (0)')
    plt.xlabel(label1)
    plt.ylabel(label2)
    plt.xticks(np.arange(0, 250, 50))
    plt.yticks(np.arange(50, 300, 50))
    plt.xlim(0, 200)
    plt.ylim(50, 250)
    plt.title(title)
    plt.legend(["Alto Riesgo (1)", "Bajo Riesgo (0)"])
    
def arterial_pressure_risk(x) -> float:
    """Función de probabilidad de riesgo en función de la presión arterial.

    Args:
        x (int): presion arterial

    Returns:
        float: probabilidad de riesgo
    """
    if (x < 100) :
        return (50 + (100-x))/70
    elif (x > 120) :
        return (x - 100) / 50
    else :
        return (x - 100) / 100
        
def colestherol_risk(x) -> float:
    """Función de probabilidad de riesgo en función del colesterol.

    Args:
        x (int): colesterol

    Returns:
        float: probabilidad de riesgo
    """
    return (x - 180) / 50
        
def gen_data(size) :
    """Función generadora de datos sintéticos para el problema de clasificación.
    Genera datos de presión arterial y colesterol, así como la clase de riesgo asociada a cada individuo.

    Args:
        size (int): tamaño poblacional de los datos a generar

    Returns:
        (array, array, array): tres arrays que representan la presión arterial, el colesterol y la clase de riesgo respectivamente.
    """
    # Seguimos distribuciones normales para centrar los datos
    xPres = np.random.normal(110, min(110-70, 180-110)-5, size)
    xCol = np.random.normal(160, min(160-120, 250-160)-5, size)
    
    # polinomio de los pesos de las probabilidades
    probRiesgo = np.vectorize(arterial_pressure_risk)(xPres) + colestherol_risk(xCol)
    # consideremos en riesgo a todas aquellas personas con una probabilidad acumulada mayor a un medio
    riesgo = (probRiesgo > 0.5).astype(int)
    
    return xPres, xCol, riesgo

def instanciate_perceptron(
    X : np.ndarray,
    y : np.ndarray,
    scaler,
    layer : tuple = (70, 30),
    alpha : float = 0.25,
    learning_rate : float = 0.25,
    test_size : float = 0.45,
    epochs : int = 5000,
) :
    """ Generamos una instancia de un clasificador perceptrón multicapa así como la partición de los datos de entrada y salida

    Args:
        X (np.ndarray): datos de entrada
        y (np.ndarray): datos de salida
        scaler (_type_): objeto de escalado de los datos
        layer (tuple, optional): neuronas por capa oculta del perceptron. Defaults to (70, 30).
        alpha (float, optional): coeficiente de regularización. Defaults to 0.25.
        learning_rate (float, optional): coeficiente inicial de aprendizaje. Defaults to 0.25.
        test_size (float, optional): porcentaje del tamaño del test. Defaults to 0.45.
        epochs (int, optional): máximo de iteraciones para el entrenamiento. Defaults to 5000.

    Returns:
        _type_: modelo entrenado, datos partidos en entrada y salida, y en entrenamiento y test
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=SEED, shuffle=False)
    
    model : MLPClassifier = MLPClassifier(layer, alpha=alpha, learning_rate_init=learning_rate, max_iter=epochs)
    
    model.fit(scaler.transform(X_train), y_train)
    
    return model, X_train, X_test, y_train, y_test
```

## Instanciación del modelo
Usaremos los siguientes datos y modelo cuando sea necesario para el resto de apartados:

```python
Xpres, Xcol, y = gen_data(POBLATION_SIZE)
X = np.array([Xpres, Xcol]).T

scaler = MinMaxScaler()
scaler.fit(X)

model_obj, X_train, X_test, y_train, y_test = instanciate_perceptron(X, y, scaler)

print(f"Precisión del modelo: {precision_score(y_test, model_obj.predict(scaler.transform(X_test)))}")
print(f"Recall del modelo: {recall_score(y_test, model_obj.predict(scaler.transform(X_test)))}")
```

```python
def indv_stats(indexInd) -> None:
    """ Función que imprime las estadísticas de un individuo en función de su índice

    Args:
        index (int): índice del individuo
    """
    print(f"Individuo seleccionado: {indexInd}")
    print(f"Presion Arterial: {X[indexInd][0]}")
    print(f"Colesterol: {X[indexInd][1]}")
    print(f"Riesgo: {(lambda x: 'Alto Riesgo (1)' if x == 1 else 'Bajo Riesgo (0)')(y[indexInd])}")
    print(f"Predicción del modelo: {(lambda x: 'Alto Riesgo (1)' if x == 1 else 'Bajo Riesgo (0)')(model_obj.predict(scaler.transform(X[indexInd].reshape(1, -1))))}")
```

```python
plot_data(Xpres, Xcol, y, "Presion Arterial", "Colesterol", "Datos Generados")
plt.show()
```

```python
plot_data(X_test[:,0], X_test[:,1], model_obj.predict(scaler.transform(X_test)), "Presion Arterial", "Colesterol", "Predicción del Modelo")
plt.show()
```

## Ejercicio 4

```python
%load_ext autoreload

%autoreload 2

import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
import dice_ml as dice

# Crear DataFrame X_pd con los datos originales
X_pd = pd.DataFrame({
    "Presion Arterial": Xpres,
    "Colesterol": Xcol,
    "Riesgo": y
}, dtype=float)

# Crear pipeline con el escalador y modelo existentes
pipeline = Pipeline([('scaler', scaler), ('classifier', model_obj)])

# Configurar DiCE
d = dice.Data(dataframe=X_pd, continuous_features=["Presion Arterial", "Colesterol"], outcome_name="Riesgo")
m = dice.Model(model=pipeline, backend="sklearn")
dice_explainer = dice.Dice(d, m, method="genetic")

# Usar el mismo individuo que en SHAP y LIME
query_instance = X_pd.iloc[individuo:individuo+1][["Presion Arterial", "Colesterol"]]

# Generar 3 contrafactuales
exp = dice_explainer.generate_counterfactuals(query_instance, total_CFs=3, desired_class="opposite")

# Mostrar contrafactuales
print("a. Descripción de los contraejemplos obtenidos con DiCE:")
print("Instancia original y contrafactuales:")
exp.visualize_as_dataframe()  # Sin argumentos adicionales
print("\nConjunto diverso de contrafactuales (mostrando solo cambios):")
exp.visualize_as_dataframe(show_only_changes=True)

# Mostrar estadísticas del individuo con tu función
print("\nEstadísticas del individuo:")
indv_stats(individuo)

# Responder a la pregunta b
print("\nb. Análisis de cómo afectan los atributos de colesterol y presión:")
cfs_df = exp.cf_examples_list[0].final_cfs_df  # Obtener DataFrame de contrafactuales
original = query_instance.iloc[0]
for i in range(3):
    cf = cfs_df.iloc[i]
    delta_presion = cf["Presion Arterial"] - original["Presion Arterial"]
    delta_colesterol = cf["Colesterol"] - original["Colesterol"]
    print(f"Contrafactual {i+1}:")
    print(f"  Cambio en Presión Arterial: {delta_presion:.4f} {'(disminuye)' if delta_presion < 0 else '(aumenta)'}")
    print(f"  Cambio en Colesterol: {delta_colesterol:.4f} {'(disminuye)' if delta_colesterol < 0 else '(aumenta)'}")

# Determinar cuál afecta más
presion_changes = [abs(cfs_df.iloc[i]["Presion Arterial"] - original["Presion Arterial"]) for i in range(3)]
colesterol_changes = [abs(cfs_df.iloc[i]["Colesterol"] - original["Colesterol"]) for i in range(3)]
avg_presion_change = np.mean(presion_changes)
avg_colesterol_change = np.mean(colesterol_changes)
print(f"\nPromedio de cambio absoluto en Presión Arterial: {avg_presion_change:.4f}")
print(f"Promedio de cambio absoluto en Colesterol: {avg_colesterol_change:.4f}")
print(f"Atributo que afecta más: {'Presión Arterial' if avg_presion_change > avg_colesterol_change else 'Colesterol'}")
```

##### Análisis

Los contraejemplos muestran que, para el individuo con presión arterial de 86.26 y colesterol de 190.60 (riesgo alto, predicción alta), se necesitan aumentos en la presión arterial (entre 8.36 y 10.50) y reducciones significativas en el colesterol (entre 25.08 y 27.10) para cambiar la predicción a bajo riesgo. Esto indica que ambos atributos son relevantes, pero el colesterol requiere ajustes más grandes.

Analisis por atributos:
- Presión Arterial: Aumenta en todos los casos (promedio de cambio absoluto: 9.28), sugiriendo que un aumento moderado puede ser necesario, pero no suficiente solo

- Colesterol: Disminuye significativamente en todos los casos (promedio de cambio absoluto: 26.24), indicando que es más crítico para cambiar la predicción.  

Conclusion: 

DiCE genera contrafactuales que muestran cómo cambiar la predicción de "Alto Riesgo" a "Bajo Riesgo" para el individuo 2728, destacando que una reducción significativa del colesterol (de 190.6041 a ~164-165, un cambio promedio de 26.2418) es esencial, mientras que un aumento moderado en la presión arterial (de 86.2626 a ~94-96, un cambio promedio de 9.2793) es tolerado por el modelo. Esto refuerza que el colesterol es el factor más influyente en la predicción, ya que los cambios requeridos en este atributo son mucho mayores y consistentes en todos los contrafactuales, sugiriendo que una intervención enfocada en reducir el colesterol sería la más efectiva para disminuir el riesgo cardiovascular en este caso.
